```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/vmajber/R---lend/master/combined2.csv')
```

```
df.head()
```

```
from sklearn.preprocessing import LabelEncoder

beso=LabelEncoder()
df['grade'] = beso.fit_transform(df['grade'])
tul=LabelEncoder()
df[ 'home_ownership'] = tul.fit_transform(df[ 'home_ownership'])
fl = LabelEncoder()
df['term'] = fl.fit_transform(df['term'].astype(str))
al=LabelEncoder()
df['addr_state'] = al.fit_transform(df['addr_state'])
purp=LabelEncoder()
df['purpose'] = purp.fit_transform(df['purpose'])
subg=LabelEncoder()
df['sub_grade'] = subg.fit_transform(df['sub_grade'])
```

```
df.head()
```

```
from google.colab import files

df.to_csv('df.csv')
files.download('df.csv')
```
